### Decision Tree Example
---

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
# Import data
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['target'] = iris.target
```

```python
# Split test / train

x_train, x_test, y_train, y_test = train_test_split(iris_df[iris.feature_names], iris_df['target'], random_state = 0, test_size = 0.2)
```

```python
# Model build & training
decision_tree = DecisionTreeClassifier(max_depth = 4, random_state = 0)
decision_tree.fit(x_train, y_train)

# Model test
decision_tree.predict(x_test)
print('Model accuracy: ', round(100 * decision_tree.score(x_test, y_test),1), '%')
```

```python
# Visualise the tree
tree.plot_tree(decision_tree);
```

```python
fn = iris.feature_names
cn = iris.target_names
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (10,8), dpi=800)
tree.plot_tree(decision_tree,
               feature_names = fn, 
               class_names = cn,
               filled = True);
fig.savefig('DecisionTree.png')
```
